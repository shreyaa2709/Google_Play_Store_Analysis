```python
import pandas as pd

# Load dataset (replace 'your_file.csv' with your file path)
df = pd.read_csv("Shreya Tyagi_GooglePlay_EDA.csv")

# Drop rows with missing 'Rating'
df = df.dropna(subset=['Rating'])

# Fill missing Content Rating with mode
df['Content Rating'] = df['Content Rating'].fillna(df['Content Rating'].mode()[0])

# Fill missing Type with "Free"
df['Type'] = df['Type'].fillna("Free")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
# Make sure all values are strings
df['Installs'] = df['Installs'].astype(str)

# Remove + and , then convert safely
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)

# Convert back to numeric
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').fillna(0).astype(int)
```

```python
# Ensure all values are strings
df['Price'] = df['Price'].astype(str)

# Remove the dollar sign
df['Price'] = df['Price'].str.replace('$', '', regex=True)

# Convert to numeric safely
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Fill invalid ones (like 'Free') with 0
df['Price'] = df['Price'].fillna(0.0)
```

```python
def convert_size(size):
    if pd.isna(size):   # handle NaN values
        return None
    size = str(size).strip()   # make sure it's a string
    
    if size == 'Varies with device':
        return None
    
    if size.endswith('M'):
        return float(size.replace('M',''))
    elif size.endswith('k'):
        return float(size.replace('k','')) / 1024   # KB → MB
    else:
        # Try to convert directly, return None if fails
        try:
            return float(size)
        except ValueError:
            return None

df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Year'] = df['Last Updated'].dt.year.astype('Int64')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: count Content Rating
gb = df['Content Rating'].value_counts()
gb = gb[gb > 0]   # ✅ remove categories with 0 count

plt.figure(figsize=(8,8))

# ✅ Custom autopct: skip 0.0%
def autopct_format(pct):
    return f'{pct:.1f}%' if pct > 0.0 else ''

# Plot donut chart
wedges, texts, autotexts = plt.pie(
    gb,
    autopct=autopct_format,
    startangle=90,
    counterclock=False,
    pctdistance=0.75,
    labeldistance=1.2
)

# Donut hole
centre_circle = plt.Circle((0,0),0.60,fc='white')
plt.gca().add_artist(centre_circle)

# Legend outside (avoid overlaps)
plt.legend(
    wedges,
    gb.index,
    title="Content Rating",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title("Distribution of Apps by Content Rating (Donut Chart)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load dataset
df = pd.read_csv("Shreya Tyagi_GooglePlay_EDA.csv")

# ------------------------------
# STEP 1: Clean column names
# ------------------------------
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('\xa0','', regex=True)

# ------------------------------
# STEP 2: Clean Installs column
# ------------------------------
df['Installs'] = df['Installs'].astype(str).str.replace('+','', regex=False).str.replace(',','', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# ------------------------------
# STEP 3: Group by Category
# ------------------------------
top_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)

# ------------------------------
# STEP 4: Plot
# ------------------------------
plt.figure(figsize=(12,6))
bars = top_installs.plot(kind='bar', color=plt.cm.tab10.colors)

# Format y-axis in millions
formatter = FuncFormatter(lambda x, _: f'{int(x/1_000_000)}M')
plt.gca().yaxis.set_major_formatter(formatter)

# Add labels on top of bars
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             f'{int(bar.get_height()/1_000_000)}M',
             ha='center', va='bottom', fontsize=10, rotation=0)

# Titles and labels
plt.title("Top 10 Categories by Installs")
plt.xlabel("Category")
plt.ylabel("Total Installs (in Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
df['Rating'].dropna().value_counts().sort_index().plot(
    kind='bar', color="cornflowerblue", edgecolor="black"
)
plt.title("Distribution Of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.xticks(rotation=0)
plt.show()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

top_reviews = df[['App','Reviews']].sort_values(by='Reviews', ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.barh(top_reviews['App'], top_reviews['Reviews'], color="skyblue")
plt.title("Top 4 Apps by Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("App")
plt.gca().invert_yaxis()  # highest on top
plt.show()
```

```python
cat_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(12,6))
cat_rating.plot(kind='bar', color="pink")
plt.title("Top 15 Categories by Average Rating")
plt.ylabel("Average Rating")
plt.ylim(3.5,5)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[df['Type']=="Paid"]['Price'].str.replace('$','').astype(float).hist(bins=40, color="orange", edgecolor="black")
plt.title("Distribution of Paid App Prices")
plt.xlabel("Price (in USD)")
plt.ylabel("Number of Apps")
plt.xlim(0,100)  # focus on apps < $100
plt.show()
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Year'] = df['Last Updated'].dt.year

apps_per_year = df.groupby('Year')['App'].count()

plt.figure(figsize=(10,5))
apps_per_year.plot(kind='line', marker='o', color="orange")
plt.title("Number of Apps Published per Year")
plt.xlabel("Year")
plt.ylabel("Number of Apps")
plt.grid(True)
plt.show()
```
